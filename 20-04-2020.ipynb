### Regular Expression
- Used for pattern matching
  ### EmailID,Mobail Number,RollNumber
- The package can be represented by 're'
- For Mathching the numbers -[0-9]
- For lower case characters - [a-z]
- For Upper case charecter - [A-z]

```python
import re
```

```python
help(re)
```

```python
# Let's define a function to match the 3 digit number
# 100 - 999 -- Valid
# less than 100 and greater than or equals to 1000 - Invalid\
def twodigitmatch(n):
    pattern = '^[0-9]{3}$'
    n= str(n)
    if re.match(pattern,n):
        return True
    return False
print(twodigitmatch(199)) #valid
print(twodigitmatch(1000))#invalid
```

```python
# Let's define a function to match the 5 lower charecter
# abcde -- Valid
# less than 5 and greater than 5 - Invalid\
def twodigitmatch(n):
    pattern = '^[a-z]{5}$'
    if re.match(pattern,n):
        return True
    return False
print(twodigitmatch('abcde')) #valid
print(twodigitmatch('abcdef'))#invalid
```

```python
# Let's define a function to match the 5 upper case
# ABCDE -- Valid
# less than 5 and greater than 5 - Invalid\
def twodigitmatch(n):
    pattern = '^[A-Z]{5}$'
    if re.match(pattern,n):
        return True
    return False
print(twodigitmatch('ABCDE')) #valid
print(twodigitmatch('ABCD'))#invalid
```
